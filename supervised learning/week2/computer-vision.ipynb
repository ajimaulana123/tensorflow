# Training Data MINST

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print(x_train, y_train)
print(x_test, y_train)
```

```python
# Normalisasi data
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

print(x_train)
print(x_test)
```

```python
# Ubah bentuk data menjadi (jumlah_sample, tinggi, lebar, channel)
x_train = x_train.reshape((60000, 28, 28, 1))
x_test = x_test.reshape((10000, 28, 28, 1))

print(x_train)
print(x_test)
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
predictions = model.predict(x_test)

plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(f'Prediksi: {predictions[i].argmax()}')
    plt.axis('off')
plt.show()
```

# Tampilkan data lebl

```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Load dataset MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Menampilkan 10 contoh gambar dengan label
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f'Label: {y_train[i]}')
    plt.axis('off')
plt.show()
```

```python
import numpy as np

# Contoh: mengubah label 0-4 menjadi satu label "Kecil" dan 5-9 menjadi "Besar"
y_custom = np.where(y_train < 5, 'Kecil', 'Besar')

# Menampilkan beberapa contoh dengan label baru
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f'Label: {y_custom[i]}')
    plt.axis('off')
plt.show()
```

```python
import pandas as pd

# Buat DataFrame dengan gambar dan label
df = pd.DataFrame({
    'Image': list(x_train.reshape(-1, 28*28)),
    'Label': y_custom
})

# Simpan DataFrame ke CSV
df.to_csv('mnist_custom_labels.csv', index=False)
```
